```python
def annotate_bar_graph(ax):
    for p in ax.patches:
        p.set_width(.6)
        current_width = p.get_width()
        diff = current_width - .35
        p.set_x(p.get_x() + diff * .5)
        ax.annotate('{0}'.format(int(float(str(p.get_height())))), (p.get_x(), p.get_height()))

ax = (application_data['TARGET'].value_counts(normalize=True)*100).plot.bar()
plt.title("Clients with late payment issue vs good clients")
plt.xlabel("0-> Good Clients; 1-> Clients having late payment issues")
plt.ylabel("Percentage")
annotate_bar_graph(ax)
plt.show()
```

The bank has late payment issues with 8% of the clients. Out of which most issues are from the clients within age group 20 and 30
